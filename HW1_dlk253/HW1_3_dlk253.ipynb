That becomes possible because computers are unable to poduce truly random numbers. When we ask a computer to generate a random number, it will in fact generate a "pseudo random" number. The number will be random in the sense that it will be some number between, for example, 0 and 1 if we ask for a **uniform random distribution** or a number near 0 if we aks for a **Gaussian distribution with mean 0**. <i> But the sequence of pseudo-random numbers that a computer generates will depend on its **seed**: if you know the seed you know all the numbers that are "randomely" drawn</i>.  

You can read more about pseudorandom number generation in this wikipedia article https://en.wikipedia.org/wiki/Pseudorandom_number_generator

## 3. now plot them: 

you can use the matplotlib library, and the function scatter(), but that would plot everything in the same color unless you ask it not to. For convenience, let's use the function plot() which by default plots lines, and tell it to plot points (in addition plot is actually more efficient, so if you have a large dataset you may want to consider using plot with points for it anyways)

https://scipy-lectures.github.io/intro/matplotlib/matplotlib.html#scatter-plots

```python
#I used this pyplot tutorial to make this: 
#http://matplotlib.org/users/pyplot_tutorial.html
#also this was useful
#http://www.scipy-lectures.org/intro/matplotlib/matplotlib.html#scatter-plots
n = 100
X = np.random.normal(0,1, n)
Y = np.random.normal(0,1, n)
T = np.arctan2(Y, X)

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.scatter(X, Y, s=50, c=T, alpha=.6)

plt.xlim(-2.5, 2.5)
plt.ylim(-2.5, 2.5)
plt.axis([-2.5, 2.5, -2.5, 2.5])
plt.margins(0.8)
plt.grid(True)
plt.title("Plotting the Distribution as Points", fontsize=18, color='blue')
plt.xlabel('X Distro of Standard Dev of 1', fontsize=14, color='blue')
plt.ylabel('Y Distro of Standard Dev of 1', fontsize=14, color='blue')


plt.show()
```
